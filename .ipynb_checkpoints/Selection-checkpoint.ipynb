```python
from sklearn.datasets import load_iris
from sklearn.linear_model import SGDClassifier
from sklearn.feature_selection import VarianceThreshold, SelectKBest, chi2, SelectFromModel, RFE, RFECV
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data
y = iris.target

plt.figure(figsize=(12,8))
plt.plot(X)
plt.legend(iris.feature_names)
```

```python
X.var(axis=0)
```

```python
selector = VarianceThreshold(threshold=0.2)
selector.fit_transform(X)
selector.get_support()
```

```python
np.array(iris.feature_names)[selector.get_support()]
```

```python
chi2(X, y)
```

```python
selector = SelectKBest(chi2, k=1)
selector.fit_transform(X, y)
selector.get_support()
```

```python
select = SelectFromModel(SGDClassifier(random_state=0), threshold='mean')
select.fit_transform(X, y)
select.get_support()
select.estimator_.coef_.mean(axis=0).mean()
```

```python
selector = RFECV(SGDClassifier(random_state=0), step=1, min_features_to_select=2, cv=4)
selector.fit(X, y)
selector.ranking_
selector.grid_scores_
```
