```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

m = 100
X = np.linspace(0,4,m).reshape((m,1))
y = 2 + X**1.3 + np.random.randn(m,1)
y = y.ravel()

plt.scatter(X, y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_pred, c='r', lw=3)
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.scatter(X, y)
plt.plot(X, y*0.8, ls='--', c='g')
plt.plot(X, y*1.2, ls='--', c='g')
plt.plot(X, y_pred, c='r', lw=3)
plt.xlabel('x')
plt.ylabel('y')
```

```python
def custom_metric(y, y_pred):
    return np.sum((y_pred>y*0.8) & (y_pred<y*1.2))/y.size
```

```python
custom_metric(y, y_pred)
```

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
custom_score = make_scorer(custom_metric, greater_is_better=True)
cross_val_score(LinearRegression(), X, y, cv=3, scoring=custom_score)
```

```python
from sklearn.svm import SVR

model = SVR(kernel='rbf', degree=3)
params = {'gamma' : np.arange(0.1,1,0.01)}
grid = GridSearchCV(model, params, cv=3, scoring=custom_score)
grid.fit(X, y)
```

```python
best_model = grid.best_estimator_
y_pred = best_model.predict(X)
custom_metric(y, y_pred)
```

```python
plt.scatter(X, y)
plt.plot(X, y*0.8, ls='--', c='g')
plt.plot(X, y*1.2, ls='--', c='g')
plt.plot(X, y_pred, c='r', lw=3)
plt.xlabel('x')
plt.ylabel('y')
```
